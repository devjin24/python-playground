# 퍼셉트론

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2 * w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1

print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

```python
import numpy as np
x = np.array([0,1])
w = np.array([0.5,0.5])
b = -0.7
print(w*x)
print(np.sum(w*x))
print(np.sum(w*x) + b)
```

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    elif tmp > 0:
        return 1


print(AND(0, 0))
print(AND(1, 0))
print(AND(0, 1))
print(AND(1, 1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -0.5 + x1 + x2 = 0
x1 = np.linspace(-2, 2, 100)
x2 = -x1 + 0.5

plt.figure(figsize=(8, 6))

# 결정 경계선 그리기
plt.plot(x1, x2, 'r--', label='Decision Boundary')

points_x = [0, 1, 0, 1]
points_y = [0, 0, 1, 1]
plt.scatter(points_x, points_y, color='red', s=100, label='Points')


# 축 설정
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.grid(True)
plt.xlabel('x₁')
plt.ylabel('x₂')

# 범례
plt.legend()

# 제목
plt.title('Decision Boundary: -0.5 + x₁ + x₂ = 0')

plt.show()
```

```python
# x1 데이터 포인트 생성
x1 = np.linspace(-2, 2, 100)
y1 = -x1 + 0.5
y2 = -x1 + 1.5

plt.figure(figsize=(8, 6))

# 결정 경계선 그리기
plt.plot(x1, y1, 'r--', label='OR Decision Boundary')
plt.plot(x1, y2, 'b--', label='NAND Decision Boundary')

# XOR 점들 
points_x = [0, 1, 0, 1]
points_y = [0, 0, 1, 1]
colors = ['blue', 'red', 'red', 'blue']  # XOR 출력값에 따라 색상 지정
plt.scatter(points_x, points_y, c=colors, s=100, label='Points')

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.grid(True)
plt.xlabel('x₁')
plt.ylabel('x₂')
plt.legend()
plt.title('XOR Decision Boundaries')

plt.show()
```
